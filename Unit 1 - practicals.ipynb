```python
from nltk.tokenize import word_tokenize, sent_tokenize

text = "Redmi has always have been the the king of budget segment.\
And yet another they gave us the Redmi 6."
word_tokens = word_tokenize(text)
sent_tokens = sent_tokenize(text)

print(f"Word tokens are:{word_tokens} respectfully")
print("*"*100)
print(f"Sentence tokens are:{sent_tokens} respectfully")
```

```python
import string
text = "OMG!!! This is the BEST product I've purchased. Highly recommend!!!"
word_tokens = word_tokenize(text)
clean_tokens = []
for word in word_tokens:
    if word not in string.punctuation:
        clean_tokens.append(word)
print(clean_tokens)
print(string.punctuation)
```

```python
#Punctuation removal - another way
import string
#input text
text = "OMG!!! This is the BEST product I've purchased. Highly recommend!!!"
#tokenize the text into words
word_tokens = word_tokenize(text)
#cleaning the text - removing punctuations
clean_text = text.translate(str.maketrans('','', string.punctuation))
print(clean_text)
```

```python
from nltk.corpus import stopwords

stopwords = set(stopwords.words('english'))
# print(stopwords)

text = "OMG!!! This is the BEST product I've purchased. Highly recommend!!!"
word_tokens = word_tokenize(text)
clean_tokens = []
for word in word_tokens:
    if word not in stopwords:
        clean_tokens.append(word)
print(clean_tokens)
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

reviews_text = """
The product quality is amazing. Fast delivery and excellent service.
Great product. Good quality. Highly recommend.
Excellent quality. Fast shipping. Love the product.
"""
word_tokens = word_tokenize(reviews_text)

stemmed_tokens = [stemmer.stem(word) for word in word_tokens]
# print(stemmed_tokens)

lemmatized_tokens = [lemmatizer.lemmatize(word, pos='v') for word in word_tokens]
print(lemmatized_tokens)
```
